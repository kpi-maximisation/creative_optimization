```python
import pandas as pd
rootdir = '../data/Challenge_Data/Assets/'
max_dir=10
dir_count=0

logo_counters = 0
l=0
max_loop=1
loop=0
p = []
for subdir, dirs, files in os.walk(rootdir):
    loop+=1
    if loop > max_loop:
        break # In order just to complete only one walkthrough
    for dir in dirs:
        dir_count+=1
        # if dir_count > max_dir:
        #     break
        dir_path = os.path.join(subdir, dir)
        # dir_path = "../data/Challenge_Data/Assets/2a6bbf47edb9f1cb3a722784c5e73e83"
        logo_loc_dict = {}
        for _,d, contents in os.walk(dir_path):
            # print(len(contents))
            # p.append(contents)
            logo_file_name = ""
            for content in contents:
                # Loop through all the contents 
                if (str(content))[-3:] == "png" or (str(content))[-3:] == "jpg":
                    # check if the current content is cta file
                    # content_name = str(content).split(".")
                    if  "logo" in str(content):
                        logo_file_name = content
                        l+=1
                        break
                    else:
                        continue
                    # p.append(content)
                    # print(colors_x)
                    # break
                # print(content) start_frame
            start_frame_dir = "../data/end_beg_frames/start_frame/"
            for _, _, start_frame_files in os.walk(start_frame_dir):
                # print(f"LEN: l{len(start_frame_files)}")
                for f in start_frame_files:
                    if dir in str(f):
                        if logo_file_name:
                            # print("FILENAME")
                            # print(logo_file_name)
                            # print(content)
                            try:
                                vals = locate_image_on_image(f"{dir_path}/{logo_file_name}", f"{start_frame_dir}/{f}")
                                if vals is not None:
                                    logo_loc_dict = {
                                        "game_id":dir,
                                        "logo_start_x":vals[2][0],
                                        "logo_start_y":vals[2][1],
                                        "logo_end_x":vals[3][0],
                                        "logo_end_y":vals[3][1],
                                        "logo_height":vals[0],
                                        "logo_width":vals[1]
                                    }
                                # print("$$$$$$$$")
                                # print(vals)
                            except Exception as e:
                                print(str(e))
                        break
                    else:
                        continue
                break
            if logo_loc_dict:
                break
            
            
            for content in contents:
                if (str(content))[-3:] == "png" or (str(content))[-3:] == "jpg":
                    # print(content)
                    if "_preview" in str(content):
                        # print("FOUND")
                        # print(logo_file_name)
                        if logo_file_name:
                            # print("FILENAME")
                            # print(logo_file_name)
                            # print(content)
                            try:
                                vals = locate_image_on_image(f"{dir_path}/{logo_file_name}", f"{dir_path}/{content}")
                                if vals is not None:
                                    logo_loc_dict = {
                                        "game_id":dir,
                                        "logo_start_x":vals[2][0],
                                        "logo_start_y":vals[2][1],
                                        "logo_end_x":vals[3][0],
                                        "logo_end_y":vals[3][1],
                                        "logo_height":vals[0],
                                        "logo_width":vals[1]
                                    }
                                # print("$$$$$$$$")
                                # print(vals)
                            except Exception as e:
                                print(str(e))
                        break
                    # else:
                    #     print("DM")
            break
        
        # print(dir_path)
        # print(logo_loc_dict)
        if logo_loc_dict:
            # print("SAVING")
            logo_counters+=1
            df = pd.DataFrame(
                logo_loc_dict, index=[0]
            )
            # print(df.shape)   
            df.to_csv(f"{dir_path}/logo_position.csv")
        # print(df.shape)
        else:
            df = pd.DataFrame({
                "game_id":dir,
                "logo_start_x":"",
                "logo_start_y":"",
                "logo_end_x":"",
                "logo_end_y":"",
                "logo_height":"",
                "logo_width":""
            }, index=[0])
            df.to_csv(f"{dir_path}/logo_position.csv")
        # break
print("Number of logos")
print(logo_counters)
print(l)
```
