```python
import pandas as pd
import sys,os
```

#### Concat all the extracted dominant colors

```python
rootdir = '../data/Challenge_Data/Assets/'
max_loop=1
loop=0
dominant_color_merged = pd.DataFrame()
for subdir, dirs, files in os.walk(rootdir):
    loop+=1
    if loop > max_loop:
        break
    for dir in dirs:
        dir_path = os.path.join(subdir, dir)
        df = pd.read_csv(f"{dir_path}/dominant_color.csv")
        dominant_color_merged = pd.concat([dominant_color_merged, df], ignore_index=True)

dominant_color_merged.shape

        
```
